```python
n=3
if n%2==0: #if n is even
    square_sum = ((n//2)*n)**2 #n/2 pairs adding to n+1
else: #if n odd
    square_sum = ((n//2+1)*n)**2 #n/2 pairs adding to n with n at the end
    
print(square_sum)
```
