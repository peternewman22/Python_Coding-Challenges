```python
num = str(3675356291)

n = 5
k = 10

candidates = []
for i in range(k-n+1):
    temp = list(num[i:k-n+i])
    #print(temp)
    if '0' in temp:
        candidates.append(0)
    else:
        total = 1
        for x in temp:
            total = total*int(x)
        candidates.append(total)
    
print(max(candidates))
    
            
```
